```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv('https://cocl.us/datascience_survey_data')

print('Data downloaded and read into a dataframe!')
```

```python
data_df.rename(columns = {'Unnamed: 0': 'Topics'}, inplace = True)
```

```python
data_df['Total'] = data_df.sum(axis=1)
data_df
```

```python
data_df.sort_values('Very interested', ascending=False, inplace=True)
data_df
```

```python
data_df.drop(['Total'], axis=1, inplace=True)
data_df
```

```python
data_df['Very interested'] = (100*data_df['Very interested']/2233).round(2)

data_df['Somewhat interested'] = (100*data_df['Somewhat interested']/2233).round(2)

data_df['Not interested'] = (100*data_df['Not interested']/2233).round(2)
```

```python
data_df
```

```python
data_df.set_index("Topics", inplace=True)
data_df
```

```python
ax = data_df.plot(kind='bar',
             figsize = (20,8),
             width = 0.8,
             color = ( '#5cb85c', '#5bc0de', '#d9534f'),
             fontsize = 14
            )

ax.set_title("Percentage of Respondents' Interest in Data Science", fontsize = 16)

#ax.set_xlabel(data_df['Topics'])

ax.legend(fontsize = 14)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Adding the percentage values    
for p in ax.patches:
    ax.annotate("{:}".format(p.get_height()),
                xy=( p.get_x()+0.03, p.get_height()+0.5)
               )
```
